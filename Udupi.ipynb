```python
import geopandas as gpd
from shapely.geometry import Point, Polygon

# Load the GeoJSON of India from the file
india_gdf = gpd.read_file("data/geoJSON/india-pc.geojson")

# Define the boundary of Udupi, Karnataka as a Polygon
udupi_boundary = Polygon([
  (74.73205, 13.33159),  # Southwest corner
    (74.8000, 13.3200),  # Southeast corner (assuming a rectangular boundary)
    (74.8000, 13.3800),  # Northeast corner
    (74.7200, 13.3800),  # Northwest corner
])

# Convert the Polygon to a GeoSeries
udupi_series = gpd.GeoSeries(udupi_boundary)

# Set the coordinate reference system (CRS) for the Udupi series
udupi_series.crs = "EPSG:4326"  # WGS 84

# Perform a spatial operation to extract the Udupi boundary
udupi_gdf = india_gdf[india_gdf.intersects(udupi_series.unary_union)]

# Reset the index if necessary
udupi_gdf = udupi_gdf.reset_index(drop=True)

udupi_gdf.to_file("udupi_district.geojson", driver="GeoJSON")
```

```python
import geopandas as gpd
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import geopy
from geopy.geocoders import Nominatim

# Load the Udupi district boundary GeoJSON file
udupi_boundary = gpd.read_file("udupi_district.geojson")

# Load your crime dataset (replace with the actual path to your CSV file)
# Use the first 250 rows for verification
crime_data = pd.read_csv("Data/DatsSet/UdupiCrimeData_TranslatedCleaned_KannadaEnglish.csv", encoding='utf-8', nrows=250)

# Initialize a geocoder
geolocator = Nominatim(user_agent="udupi_geocoder")

# Geocode the 'Location' column to obtain latitude and longitude
def geocode_location(location):
    location_info = geolocator.geocode(location)
    if location_info:
        return location_info.latitude, location_info.longitude
    else:
        return None

# Apply geocoding to the 'Location' column
crime_data['Coordinates'] = crime_data['Location'].apply(geocode_location)

# Remove rows with missing coordinates
crime_data = crime_data.dropna(subset=['Coordinates'])

# Create a map centered on Udupi
udupi_map = folium.Map(location=[13.3417, 74.7477], zoom_start=12)  # Adjust the coordinates as needed

# Create a MarkerCluster for the crime data
marker_cluster = MarkerCluster().add_to(udupi_map)

# Add markers for each crime incident
for index, row in crime_data.iterrows():
    folium.Marker(location=row['Coordinates'], popup=row['Crime Type']).add_to(marker_cluster)

# Save the map as an HTML file or display it
udupi_map.save("udupi_crime_markers.html")

print("Markers for the first 250 crime incidents in Udupi district created. Open the HTML file to view the map.")
```

```python
import geopandas as gpd
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import geopy
from geopy.geocoders import Nominatim

udupi_boundary = gpd.read_file("udupi_district.geojson")

crime_data = pd.read_csv("Data/DatsSet/UdupiCrimeData_TranslatedCleaned_KannadaEnglish.csv", encoding='utf-8', nrows=25)

# Initialize a geocoder
geolocator = Nominatim(user_agent="udupi_geocoder")

# Geocode the 'Location' column to obtain latitude and longitude
def geocode_location(location):
    location_info = geolocator.geocode(location)
    if location_info:
        return location_info.latitude, location_info.longitude
    else:
        return None

# Apply geocoding to the 'Location' column
crime_data['Coordinates'] = crime_data['Location'].apply(geocode_location)

# Remove rows with missing coordinates
crime_data = crime_data.dropna(subset=['Coordinates'])

# Create a map centered on Udupi
udupi_map = folium.Map(location=[13.3417, 74.7477], zoom_start=12)  # Adjust the coordinates as needed

# Create a MarkerCluster for the crime data
marker_cluster = MarkerCluster().add_to(udupi_map)

# Add markers for each crime incident
for index, row in crime_data.iterrows():
    folium.Marker(location=row['Coordinates'], popup=row['Crime Type']).add_to(marker_cluster)

# Add the Udupi district boundaries
folium.GeoJson(udupi_boundary, name="Udupi District Boundaries").add_to(udupi_map)

# Save the map as an HTML file or display it
udupi_map.save("udupi_crime_markers.html")

print("Markers for the first 25 crime incidents in Udupi district, along with district boundaries, have been created. Open the HTML file to view the map.")
```
