```python
import pandas as pd

# Assuming your dataset is in a CSV file
file_path = "UdupiCrimeData_TranslatedCleaned_KannadaEnglish.csv" 

# Read the dataset into a pandas DataFrame
df = pd.read_csv(file_path)

# Check unique values in the "Crime Type" column
unique_crime_types = df["Crime Type"].unique()

# Print the unique crime types
print("Unique Crime Types:")
for crime_type in unique_crime_types:
    print(crime_type)

# Calculate the total count for each unique crime type
crime_type_counts = df["Crime Type"].value_counts()

# Print the total count for each crime type
print("\nTotal Count for Each Crime Type:")
print(crime_type_counts)
```
